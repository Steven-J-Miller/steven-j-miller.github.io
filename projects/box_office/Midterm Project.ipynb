## Replace Headers

## Format Data to a Readable Format

We currently have several columns where data is represented in JSON format. This was lost when importing our CSV and is now just a large string. We'll move it into a dictionary object and then extract the values from the dictionary for further work. I don't expect the keys to be particularly useful for this project and am okay with moving forward without them.

```python
from ast import literal_eval
cols_to_translate = ['Keywords','Genres', 'Production Companies', 'Production Countries', 'Spoken Languages']

for col in cols_to_translate:
    df[col] = df[col].apply(literal_eval)
    

for col in cols_to_translate:
    new_col = []
    for i in range(len(df)):
        new_list = [x['name'] for x in df.loc[i,col]]
        new_col.append(new_list)
    df[col] = new_col
```

## Conduct Fuzzy Matching

```python
from fuzzywuzzy import fuzz, process
import numpy as np

fuzziness = np.zeros((14,14))

all_lower =  df['Overview'].map(lambda x: str(x).lower())
all_lower = all_lower.str.replace(r'[^\w\s]+', '')

for i in range(14):
    for j in range(14):
        fuzziness[i,j] = fuzz.token_sort_ratio(all_lower[i],all_lower[j])
print(fuzziness)        
```

```python
similar = process.extract(all_lower[1359],all_lower, limit = 10, scorer=fuzz.token_sort_ratio)

# Films with plots similar to the 1989 Batman
for film in similar:
    print(df.loc[film[2],'Original Title'], film[1])
```

```python
similar = process.extract(all_lower[4267],all_lower, limit = 10, scorer=fuzz.token_sort_ratio)

# Films with plots similar to the 1966 Batman
for film in similar:
    print(df.loc[film[2],'Original Title'], film[1])
```
